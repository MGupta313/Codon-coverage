```python
import pandas as pd

pd.set_option("display.max_rows", None, "display.max_columns", None)

path = r'/Users/mansi/Desktop/Haiti/Pass_fail/Haiti_PetValues.xlsx'

df = pd.read_excel(path, sheet_name = "Combined")

list2 = list(df["dhfr"])
# print(list2)

sequencing =[]

for i in range(len(list2)):
    if list2[i] >= 0:
        sequencing.append('pass')
    else:
        sequencing.append('fail')       
# print(sequencing)

df_new = df.drop(columns = ['Sequences_name', 'mdr1', 'dhps', 'crt'])
# df_new

df_new["sequencing"] = sequencing

df_comb = df_new.rename(columns={'SPECIES_CT': 'parasites per uL'})
df_comb

import seaborn as sns
import matplotlib.pyplot as plt


my_pal = {"fail":"#ef5c00", "pass":"#006cef"}
sns.set_style("whitegrid")

g= sns.catplot(x="method", y="parasites per uL", hue="sequencing", kind="swarm", data=df_comb, palette=my_pal)
g.fig.suptitle("DHFR")
plt.locator_params(axis='y', nbins=15)
g.set(yticklabels=["","30,000","10,000","3,000", "600","150","40","10","3","1", "0.4"])
g.fig.savefig('/Users/mansi/Desktop/Haiti/Pass_fail/DHFR-PassFail.pdf', bbox_inches='tight')
```
