# Categorizing samples based on 

```python
import os
import re
import pandas as pd
# pd.set_option("display.max_rows", None, "display.max_columns", None)


"""
Categorizing Haiti individual samples based on their coverage graphs and linking them to their CT values
(~parasitemia levels).

Total samples for Haiti study = 100

Separating samples based on quality.

Key:

-3 ==> No data available. No coverage calculation was done.
-2 ==> Negligible data available. Sequencing error.
-1 ==> Less number of codons covered for respective genes.
 0 ==> Mediocre number of codons covered for respective genes.
 1 ==> Maximum or near to maximum number of codons covered for respective genes.
"""


excel_path = r'/Users/mansi/Desktop/Haiti/Individual/Codon_coverage/Haiti_PetValues.xlsx'

### Getting PET values data from tyhe excel file. Selecting just the sequences name.

df1 = pd.read_excel(excel_path, sheet_name = "CovValidation")
seq_name_list = list(df1.Sequences_name)

### Replacing Nan values to zero so that data frame can be filled with key values.

df2 = pd.read_excel(excel_path, sheet_name = "CovValidation", index_col=0)
df2.fillna(0, inplace=True)

### Getting samples for PfDHFR coverage results prepared to categorize the codon positions

def dhfr_values():

    filepath_list = []
    sample_list = []
    cov_directory = r'/Users/mansi/Desktop/Haiti/Individual/Codon_coverage/PfDHFR/Category_results/avg_cat_cov/'
    for filename in os.listdir(cov_directory):
        if filename != ".DS_Store" and filename != "test":
            filepath = os.path.join(cov_directory, filename)
            filepath_list.append(filepath)
            sample_num = int(filename[4:10].split("P")[0]) #Getting sample number/id
            sample_list.append(sample_num)

    """
    Total samples for Haiti study = 100
    Samples generated from codon coverage calculation for PfDHFR = 91
    Max number of codons for PfDHFR = 609

    Key for catrgorizing:

    -3 = No sample data
    -2 = num of codons < 100
    -1 = 100=< num of codons < 500
     0 = 500=< num of codons < 600
     1 = num of codons >= 600
    """

    no_data_samples = []
    data_samples = []
    for seq_name in seq_name_list:
        if seq_name not in sample_list:
            no_data_samples.append(seq_name)
            df2.at[seq_name, 'dhfr'] = -3 #No data for these samples. No images, nothing.
        if seq_name in sample_list:
            for sample_name in sample_list:
                if seq_name == sample_name:
                    data_samples.append(seq_name)


    for seq_id in data_samples:
        for file in filepath_list:
            file_id = int(file[71:77].split("P")[0])
            if seq_id == file_id:
                with open(file) as f:
    #                 print("filename", file)
                    lines = f.readlines()
                    for i in range(len(lines)):
                        line = lines[i].split(',')
                        num_of_codons = len(line) - 1
    #                     print("num of codons", num_of_codons)
                        if num_of_codons >= 600: ## these values will change for each gene
                            df2.at[seq_id, 'dhfr'] = 1
                        elif num_of_codons >= 500 and num_of_codons < 600:## these values will change for each gene
                            df2.at[seq_id, 'dhfr'] = 0
                        elif num_of_codons >= 100 and num_of_codons < 500:## these values will change for each gene
                            df2.at[seq_id, 'dhfr'] = -1
                        elif num_of_codons < 100:## these values will change for each gene
                            df2.at[seq_id, 'dhfr'] = -2
    return df2

dhfr_df = dhfr_values()


def mdr1_values(input_dhfr_df):
    
    filepath_list = []
    sample_list = []
    cov_directory = r'/Users/mansi/Desktop/Haiti/Individual/Codon_coverage/PfMDR1/Category_results/avg_cat_cov/'
    for filename in os.listdir(cov_directory):
        if filename != ".DS_Store": #and filename != "test":
            filepath = os.path.join(cov_directory, filename)
            filepath_list.append(filepath)
            sample_num = int(filename[4:10].split("P")[0]) #Getting sample number/id
            sample_list.append(sample_num)

    """
    Total samples for Haiti study = 100
    Samples generated from codon coverage calculation for PfMDR1 = 85
    Max number of codons for PfMDR1 = 1404

    Key for catergorizing:

    -3 = No sample data
    -2 = num of codons < 140
    -1 = 140=< num of codons < 1050
     0 = 1050=< num of codons < 1330
     1 = num of codons >= 1330
    """

    no_data_samples = []
    data_samples = []
    for seq_name in seq_name_list:
        if seq_name not in sample_list:
            no_data_samples.append(seq_name)
            df2.at[seq_name, 'mdr1'] = -3 #No data for these samples. No images, nothing.
        if seq_name in sample_list:
            for sample_name in sample_list:
                if seq_name == sample_name:
                    data_samples.append(seq_name)


    for seq_id in data_samples:
        for file in filepath_list:
            file_id = int(file[71:77].split("P")[0])
            if seq_id == file_id:
                with open(file) as f:
    #                 print("filename", file)
                    lines = f.readlines()
                    for i in range(len(lines)):
                        line = lines[i].split(',')
                        num_of_codons = len(line) - 1
    #                     print("num of codons", num_of_codons)
                        if num_of_codons >= 1330: ## these values will change for each gene
                            input_dhfr_df.at[seq_id, 'mdr1'] = 1
                        elif num_of_codons >= 1050 and num_of_codons < 1330:## these values will change for each gene
                            input_dhfr_df.at[seq_id, 'mdr1'] = 0
                        elif num_of_codons >= 140 and num_of_codons < 1050:## these values will change for each gene
                            input_dhfr_df.at[seq_id, 'mdr1'] = -1
                        elif num_of_codons < 140:## these values will change for each gene
                            input_dhfr_df.at[seq_id, 'mdr1'] = -2

    return input_dhfr_df

mdr1_df = mdr1_values(dhfr_df)


def dhps_values(input_mdr1_df):

    filepath_list = []
    sample_list = []
    cov_directory = r'/Users/mansi/Desktop/Haiti/Individual/Codon_coverage/PfDHPS/Category_results/avg_cat_cov/'
    for filename in os.listdir(cov_directory):
        if filename != ".DS_Store": #and filename != "test":
            filepath = os.path.join(cov_directory, filename)
            filepath_list.append(filepath)
            sample_num = int(filename[4:10].split("P")[0]) #Getting sample number/id
            sample_list.append(sample_num)

    """
    Total samples for Haiti study = 100
    Samples generated from codon coverage calculation for PfDHPS = 84
    Max number of codons for PfDHPS = 805

    Key for catrgorizing:

    -3 = No sample data
    -2 = num of codons < 300
    -1 = 300=< num of codons < 540
     0 = 540=< num of codons < 750
     1 = num of codons >= 750
    """

    no_data_samples = []
    data_samples = []
    for seq_name in seq_name_list:
        if seq_name not in sample_list:
            no_data_samples.append(seq_name)
            df2.at[seq_name, 'dhps'] = -3 #No data for these samples. No images, nothing.
        if seq_name in sample_list:
            for sample_name in sample_list:
                if seq_name == sample_name:
                    data_samples.append(seq_name)


    for seq_id in data_samples:
        for file in filepath_list:
            file_id = int(file[71:77].split("P")[0])
            if seq_id == file_id:
                with open(file) as f:
    #                 print("filename", file)
                    lines = f.readlines()
                    for i in range(len(lines)):
                        line = lines[i].split(',')
                        num_of_codons = len(line) - 1
    #                     print("num of codons", num_of_codons)
                        if num_of_codons >= 750: ## these values will change for each gene
                            input_mdr1_df.at[seq_id, 'dhps'] = 1
                        elif num_of_codons >= 540 and num_of_codons < 750:## these values will change for each gene
                            input_mdr1_df.at[seq_id, 'dhps'] = 0
                        elif num_of_codons >= 300 and num_of_codons < 540:## these values will change for each gene
                            input_mdr1_df.at[seq_id, 'dhps'] = -1
                        elif num_of_codons < 300:## these values will change for each gene
                            input_mdr1_df.at[seq_id, 'dhps'] = -2
                            
    return input_mdr1_df

dhps_df = dhps_values(mdr1_df)

def crt_values(input_dhps_df):
    filepath_list = []
    sample_list = []
    cov_directory = r'/Users/mansi/Desktop/Haiti/Individual/Codon_coverage/PfDHPS/Category_results/avg_cat_cov/'
    for filename in os.listdir(cov_directory):
        if filename != ".DS_Store": #and filename != "test":
            filepath = os.path.join(cov_directory, filename)
            filepath_list.append(filepath)
            sample_num = int(filename[4:10].split("P")[0]) #Getting sample number/id
            sample_list.append(sample_num)

    """
    Total samples for Haiti study = 100
    Samples generated from codon coverage calculation for PfCRT = 85
    Max number of codons for PfCRT = 657
    NOTE: There is only one sample with max number of codons. Else it is near 480 codons.

    Key for catrgorizing:

    -3 = No sample data
    -2 = num of codons < 180
    -1 = 180=< num of codons < 330
     0 = 330=< num of codons < 480
     1 = num of codons >= 480
    """

    no_data_samples = []
    data_samples = []
    for seq_name in seq_name_list:
        if seq_name not in sample_list:
            no_data_samples.append(seq_name)
            df2.at[seq_name, 'crt'] = -3 #No data for these samples. No images, nothing.
        if seq_name in sample_list:
            for sample_name in sample_list:
                if seq_name == sample_name:
                    data_samples.append(seq_name)


    for seq_id in data_samples:
        for file in filepath_list:
            file_id = int(file[71:77].split("P")[0])
            if seq_id == file_id:
                with open(file) as f:
    #                 print("filename", file)
                    lines = f.readlines()
                    for i in range(len(lines)):
                        line = lines[i].split(',')
                        num_of_codons = len(line) - 1
    #                     print("num of codons", num_of_codons)
                        if num_of_codons >= 480: ## these values will change for each gene
                            input_dhps_df.at[seq_id, 'crt'] = 1
                        elif num_of_codons >= 330 and num_of_codons < 480:## these values will change for each gene
                            input_dhps_df.at[seq_id, 'crt'] = 0
                        elif num_of_codons >= 180 and num_of_codons < 330:## these values will change for each gene
                            input_dhps_df.at[seq_id, 'crt'] = -1
                        elif num_of_codons < 180:## these values will change for each gene
                            input_dhps_df.at[seq_id, 'crt'] = -2
    return input_dhps_df

crt_df = crt_values(dhps_df)
print(crt_df)
```
